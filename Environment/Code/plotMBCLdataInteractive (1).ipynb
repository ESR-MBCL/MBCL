## Example notebook to plot MBCL data

```python
# import libraries

import pandas as pd 
import xarray as xr
import os as os

import hvplot.pandas # noqa: adds hvplot method to pandas objects
import hvplot.xarray # noqa: adds hvplot method to xarray objects

```

```python
# read in the data

# location of data
fileLoc = 'C:\Data'
fileNames = os.listdir(fileLoc)
fileNames.sort()
#fileNames = [f for f in fileNames if 'md' not in f]
print(fileNames)
# colNames = [c[:-4] for f in fileNames]
os.chdir(fileLoc)

# read the file
#df = pd.read_csv("MBCP2019-2020_iButton12.csv")
df1 = pd.read_csv(fileNames[0],header= 13)
df2 = pd.read_csv(fileNames[1],header= 13)

df3 = pd.read_csv(fileNames[2],header= 13)

df4 = pd.read_csv(fileNames[3],header= 13)
df5 = pd.read_csv(fileNames[4],header= 13)
#df6 = pd.read_csv(fileNames[5],header= 13)


# create datetime variable, set as index
df1['datetime'] = pd.to_datetime(df1['Date/Time'])
df1 = df1.set_index('datetime')
df1.rename(columns = {'Value':'temperature'},inplace = True)

# drop useless columns
df1 = df1.drop(['Unit','Date/Time'],axis = 'columns')

df2['datetime'] = pd.to_datetime(df2['Date/Time'])
df2 = df2.set_index('datetime')
df2.rename(columns = {'Value':'temperature'},inplace = True)

# drop useless columns
df2 = df2.drop(['Unit','Date/Time'],axis = 'columns')

df3['datetime'] = pd.to_datetime(df3['Date/Time'])
df3 = df3.set_index('datetime')
df3.rename(columns = {'Value':'temperature'},inplace = True)

# drop useless columns
df3 = df3.drop(['Unit','Date/Time'],axis = 'columns')


df4['datetime'] = pd.to_datetime(df4['Date/Time'])
df4 = df4.set_index('datetime')
df4.rename(columns = {'Value':'temperature'},inplace = True)

# drop useless columns
df4 = df4.drop(['Unit','Date/Time'],axis = 'columns')
"""
df5['datetime'] = pd.to_datetime(df5['Date/Time'])
df5 = df5.set_index('datetime')
df5.rename(columns = {'Value':'temperature'},inplace = True)

# drop useless columns
df5 = df5.drop(['Unit','Date/Time'],axis = 'columns')


df6['datetime'] = pd.to_datetime(df6['Date/Time'])
df6 = df6.set_index('datetime')
df6.rename(columns = {'Value':'temperature'},inplace = True)

# drop useless columns
df6 = df6.drop(['Unit','Date/Time'],axis = 'columns')
"""
```

```python
dfs = [df1, df2, df3, df4, df5]

dffinal = pd.concat(dfs)
temp = dffinal['temperature']
threshold = []
line = []

for x in temp:
    if x >= 35:
        threshold.append('reached')
    else:
        threshold.append('below')
    line.append(30)

dffinal['threshold'] = threshold
dffinal['line'] = line
dffinal
```

```python
# plot the data using hvplot

plot = dffinal.hvplot.line(x='datetime',y='temperature')
#plot = dffinal.hvplot.scatter(x='datetime',y='temperature',by='threshold') * dffinal.hvplot(x='datetime',y='line', color=['#FF0000'])
hvplot.save(plot, 'iButton.html')
plot
```
